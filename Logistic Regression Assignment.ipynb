```python
# Load Libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#Load Dataset
data = pd.read_csv("bank-full.csv",sep=";")
data.head()
```

```python
data.info()
```

```python
## One-Hot Encoding of categrical variables
data1 =pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
# Custom Binary Encoding of Binary o/p variables 
data1['default'] = np.where(data1['default'].str.contains("yes"),1,0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"),1,0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"),1,0)
data1['y'] = np.where(data1['y'].str.contains("yes"),1,0)
data1
```

```python
# Find and Replace Encoding for month categorical varaible
data1['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
       
```

```python
data_new=data1.replace(order)
data_new
```

```python
data_new.info()
```

```python
data_new.shape
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_new.loc[:,['day', 'month',
       'duration', 'campaign', 'pdays', 'previous','balance','age']]=sc.fit_transform(data_new.loc[:,['day', 'month',
       'duration', 'campaign', 'pdays', 'previous','balance','age']])
```

```python
data_new.head(5)
```

# Model Building

```python
x=data_new.drop("y",axis=1)
```

```python
y=data_new["y"]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

# Model Predictions

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df= pd.DataFrame({'actual': y,
                         'predicted_prob': classifier.predict(x)})
```

```python
y_pred_df
```

# Testing Model Accuracy

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
print (confusion_matrix)
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39129+1227)/(39129+793+4062+1282)
```

```python
 classifier.predict_proba (x)[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba (x)[:,1])

auc = roc_auc_score(y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
